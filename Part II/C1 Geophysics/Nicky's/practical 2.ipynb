```python
from math import pi
import matplotlib.pyplot as plt
import numpy as np

E = 70e9 # Young's modulus in Pa
sigma = 0.25 # Poisson's ratio
```

```python
def w(w0, x, alpha):
    return -w0*np.exp(-x/alpha)*(np.cos(x/alpha) + np.sin(x/alpha))

def w0(V0, alpha, D):
    return V0*(alpha**3)/(8*D)

def alpha(D, rho_m, rho_p, g):
    return (4*D/((rho_m - rho_p)*g))**(1/4)

def D(Te):
    return (E*(Te**3))/(12*(1 - sigma**2))

def x0(alpha):
    return alpha*3*pi/4

def xb(alpha):
    return pi*alpha

def wprime(w0, alpha, x):
    return (w0*np.exp(-x/alpha)/alpha)*(-2*np.sin(x/alpha))

```

```python
Te = 40e3 # elastic thickness in m
distance = np.linspace(0e6, 1e6, 1000)

D_1 = D(Te)
alpha_1 = alpha(D_1, 3300, 1000, 9.81)
w0_1 = w0(1, alpha_1, D_1)
w_1 = w(w0_1, distance, alpha_1)

x0_1 = x0(alpha_1)
xb_1 = xb(alpha_1)


plot = plt.plot(distance, w_1)
plt.axhline(y=0, color='r', lineityle='--')
plt.axvline(x=x0_1, color='b', linestyle='--')
# plt.axvline(x=xb_1,gzgzasdaaasdas color='g', linestyle='--')
```

```python
distance = np.linspace(0e6, 2e6, 1000)
for thickness in [0.1, 5, 40, 100]:
    _D = D(thickness*1e3)
    _alpha = alpha(_D, 3300, 1000, 9.81)
    _w0 = w0(1, _alpha, _D)
    _w = w(_w0, distance, _alpha)
    plt.plot(distance, _w, label=f"{thickness} km")
    plt.legend()
    plt.show()
```
