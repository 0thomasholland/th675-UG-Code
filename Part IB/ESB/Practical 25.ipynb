```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# import datasets/pract_25.csv 

df = pd.read_csv('datasets/pract_25.csv').T

# first row as header

new_header = df.iloc[0]
df = df[1:]
df.columns = new_header

df.head()
```

```python
# plot CaO, FeO, TiO2, K2O vs MgO, group by Source

sns.pairplot(df, vars=['MgO', 'CaO', 'FeO', 'TiO2', 'K2O'], hue='Source')
```
