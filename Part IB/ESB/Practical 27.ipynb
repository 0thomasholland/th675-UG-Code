```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# import datasets/pract_25.csv 

df = pd.read_csv('datasets/pract_27.csv').T

# first row as header

new_header = df.iloc[0]
df = df[1:]
df.columns = new_header

df.head()
```

```python
# plot SiO2	Al2O3	TiO2	FeO	MgO	CaO	Na2O	K2O	P2O5

sns.pairplot(df, vars=['SiO2', 'Al2O3', 'TiO2', 'FeO', 'MgO', 'CaO', 'Na2O', 'K2O', 'P2O5'])
```

```python
# list by highest correlation

df.corr().unstack().sort_values(ascending=False).drop_duplicates()
```

```python
# add MgO/CaO, Na2O/K2O, P2O5/SiO2 ratio

df['MgO/CaO'] = df['MgO'] / df['CaO']
df['Na2O/K2O'] = df['Na2O'] / df['K2O']
df['P2O5/SiO2'] = df['P2O5'] / df['SiO2']
df['TiO2/Al2O3'] = df['TiO2'] / df['Al2O3']

# pairplot again

sns.pairplot(df, vars=['FeO', 'MgO/CaO', 'Na2O/K2O', 'P2O5/SiO2', 'TiO2/Al2O3'])
```
