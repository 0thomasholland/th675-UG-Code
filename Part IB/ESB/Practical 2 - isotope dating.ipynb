```python
import matplotlib. pyplot as plt
from math import log
import numpy as np
```

## Stattern and Nakhla meteriorites

```python
# all data normalised to 86Sr

#y
stannern_87Sr = np.array([0.7049, 0.7006, 0.6996, 0.6994, 0.6993])
nakhla_87Sr = np.array([0.7028, 0.7026])

#x
stannern_87Rb = np.array([0.090, 0.024, 0.009, 0.006, 0.005])
nakhla_87Rb = np.array([0.015, 0.004])

stannern_names = ["Stanern glass", "Stannern tridymite/quartz", "Stannern pyroxene", "Stannern bulk sample", "Stannern feldspar"]
nakhla_names = ["Nakhla pyroxene", "Nakhla feldspar"]
```

```python
s_a,s_b = np.polyfit(stannern_87Rb, stannern_87Sr, 1)
n_a,n_b = np.polyfit(nakhla_87Rb, nakhla_87Sr, 1)
```

```python
plt.scatter(stannern_87Rb, stannern_87Sr)
plt.scatter(nakhla_87Rb, nakhla_87Sr)


plt.plot(stannern_87Rb, s_a*stannern_87Rb+s_b)
plt.plot(nakhla_87Rb, n_a*nakhla_87Rb+n_b)

plt.show()
```

```python
# time calculations
time_lambda_Rb = 1.42 * pow(10, -11)

s_time = log(s_a+1)/time_lambda_Rb/pow(10,6)

print(s_time)

n_time = log(n_a+1)/time_lambda_Rb/pow(10,6)

print(n_time)
```
