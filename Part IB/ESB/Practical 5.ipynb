```python
import numpy as np
from math import log
from matplotlib import pyplot as plt
```

```python
Sm = np.array([0.1312, 0.1324, 0.2081, 0.2112])
Nd = np.array([0.510783, 0.510819, 0.513105, 0.513199])

lambda_val = 6.54 * pow(10, -12)

grad, intercept = np.polyfit(Sm, Nd, 1)

age = log(grad+1)/lambda_val
print(age)
print(intercept)
print(grad)
```

```python
plt.plot(Sm, Nd, 'o')
plt.plot(Sm, grad*Sm + intercept)
plt.xlabel('Sm')
plt.ylabel('Nd')
plt.text(0.14, 0.513, 'y = ' + str(round(grad, 4)) + 'x + ' + str(round(intercept, 4)))
```

```python
#Sm is numpy array, use average with error propagation
Sm_avg = np.average(Sm)

# define the time values
t = np.linspace(0, 4549259532, 100)

# calculate the corresponding Nd values
Nd = intercept + Sm_avg * (np.exp(lambda_val * t) - 1)

# plot the data
plt.plot(t, Nd)
plt.xlabel('Time')
plt.ylabel('Nd')
plt.show()
```
