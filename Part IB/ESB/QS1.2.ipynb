```python
%matplotlib ipympl
import matplotlib.pyplot as plt
import numpy as np
```

```python
partition_coefficient = 2
initial_concentration = 0.3

melt_produced = np.linspace(0, 1, 100)

liquid_melt_concentration = initial_concentration/(melt_produced+partition_coefficient-melt_produced*partition_coefficient)
```

```python
plt.plot(melt_produced, liquid_melt_concentration)
plt.xlabel('Melt fraction')
plt.ylabel('Liquid melt concentration of mineral')
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')

partition_coefficient = np.concatenate((np.linspace(0.2, 1, 4), np.linspace(1.2, 2, 3)))
print(partition_coefficient)
x = melt_produced
initial_concentration = np.linspace(0, 1, 100)
y = initial_concentration
# generate meshgrid
x, y = np.meshgrid(x, y)
# calculate z coordinates for each point for each partition coefficient

for i in range(len(partition_coefficient)):
    liquid_melt_concentration = y/(x+partition_coefficient[i]-x*partition_coefficient[i])
    z = liquid_melt_concentration
    ax.plot_surface(x, y, z)


ax.set_xlabel('Melt fraction')
ax.set_ylabel('Initial concentration')
ax.set_zlabel('Liquid melt concentration of mineral')
# increase plot view size

plt.show()
```

```python
# three minerals with starting concentrations of 0.3, 0.5, 0.2, and partition coefficients of 0.3, 1, 1.2, plot a 2d stackplot of melt fraction vs liquid melt concentration for each mineral


partition_coefficient = [0.5, 1, 2]
initial_concentration = [0.3, 0.5, 0.2]

melt_produced = np.linspace(0, 1, 100)

fig = plt.figure()

x = melt_produced
y0 = initial_concentration[0]/(x+partition_coefficient[0]-x*partition_coefficient[0])
y1 = initial_concentration[1]/(x+partition_coefficient[1]-x*partition_coefficient[1])
y2 = initial_concentration[2]/(x+partition_coefficient[2]-x*partition_coefficient[2])

plt.stackplot(x, y0, y1, y2, labels=['mineral 1', 'mineral 2', 'mineral 3'])
  
plt.xlabel('Melt fraction')
plt.ylabel('Liquid melt concentration of mineral')
plt.legend()

plt.show()
```
