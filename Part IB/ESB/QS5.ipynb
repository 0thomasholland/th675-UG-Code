```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy as sp
```

```python
x_LiF = np.array([1, 0.9382, 0.8993, 0.8521, 0.8021])
T = np.array([848, 824, 807, 788, 765])

ln_x_LiF = np.log(x_LiF)
T_inv = 1/T

R = sp.constants.R
```

```python
# Linear regression
slope, intercept, r_value, p_value, std_err = sp.stats.linregress(T_inv, ln_x_LiF)

print("Slope: ", slope)
print("Intercept: ", intercept)

H_fus = -slope*R
print("H_fus: ", H_fus)

S_fus = intercept*R
print("S_fus: ", S_fus)
```

```python
plt.plot(T_inv, ln_x_LiF, 'o')
plt.xlabel('1/T')
plt.ylabel('ln(x)')
plt.show()
```

```python
temp_range = np.linspace(400, 900, 100)
x_LiF_calc = np.exp(intercept + slope*(1/temp_range))
```

```python
# plot x_LiF vs T as line

plt.plot(temp_range, x_LiF_calc)
plt.xlabel('T')
plt.ylabel('x_LiF')
```
