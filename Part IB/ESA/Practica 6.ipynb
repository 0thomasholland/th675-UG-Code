```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
sigma_1 = np.array([2.03, 1.81, 1.52, 1.35, 1.06])
sigma_3 = np.array([0.47, 0.39, 0.28, 0.25, 0.24])
P = 0.5

theta = np.array([58.5, 58.5, 59, 69, 90])

color = ['red', 'orange', 'yellow', 'green', 'blue']
sigma_1_dash = sigma_1 - P
sigma_3_dash = sigma_3 - P

centre = (sigma_1_dash + sigma_3_dash)/2
radius = (sigma_1_dash - sigma_3_dash)/2
```

```python
fig, ax = plt.subplots()
angle = np.linspace( 0 , 2 * np.pi , 150 )
# add grid every 0.01
plt.grid( linestyle = '--' )

plt.axis('equal')
for i in range( len( centre ) ):
    x = radius[i] * np.cos( angle ) + centre[i]
    y = radius[i] * np.sin( angle )
    ax.plot( x , y , color = color[i] )
    ax.plot( [ centre[i] , centre[i] + radius[i] * np.cos( 2 * theta[i] * np.pi / 180 ) ] ,
             [ 0 , radius[i] * np.sin( 2 * theta[i] * np.pi / 180 ) ] , color = color[i] )
    ax.plot( [ centre[i] , centre[i] + radius[i] * np.cos( -2 * theta[i] * np.pi / 180 ) ] ,
             [ 0 , radius[i] * np.sin( -2 * theta[i] * np.pi / 180 ) ] , color = color[i] )

# for the positive x values, plot a straight line through the failure point
positive_x = centre + radius * np.cos( 2 * theta * np.pi / 180)
positive_y = radius * np.sin( 2 * theta * np.pi / 180 )

#plot a line through the first three points, and extend it through from x=0  to x= 1.5, using formula y = mx + c from the first three points
m = (positive_y[2] - positive_y[0])/(positive_x[2] - positive_x[0])
c = positive_y[2] - m*positive_x[2]
plt.plot( [0, 1.5], [c, m*1.5 + c], color = 'black' )

# # x is range of x values from positive_x[4] to 0
# x = np.linspace(positive_x[4], 0, 100)

# # y is given by Lagrange Interpolation Formula, with x1 = positive_x[4], x2 = positive_x[3], x3 = positive_x[2], y1 = positive_y[4], y2 = positive_y[3], y3 = positive_y[2]
# x1 = - positive_x[4]
# x2 = - positive_x[3]
# x3 = 0
# y1 = positive_y[4]
# y2 = positive_y[3]
# y3 = c

# print(x1, x2, x3, y1, y2, y3)

# y = (y1*(x-x2)*(x-x3)/(x1-x2)*(x1-x3)) + (y2*(x-x1)*(x-x3)/(x2-x1)*(x2-x3)) + (y3*(x-x1)*(x-x2)/(x3-x1)*(x3-x2))


# plt.plot(x, y, color = 'black')
```
